```python
board_w_in = 48
board_h_in = 72

border_in = 2
square_in = 2

squares_x = 22
squares_y = 34

svg = []

svg.append(f'<svg xmlns="http://www.w3.org/2000/svg" width="{board_w_in}in" height="{board_h_in}in" viewBox="0 0 {board_w_in} {board_h_in}">')
svg.append(f'<rect x="0" y="0" width="{board_w_in}" height="{board_h_in}" fill="white"/>')

for y in range(squares_y):
    for x in range(squares_x):
        if (x + y) % 2 == 0:
            rx = border_in + x * square_in
            ry = border_in + y * square_in
            svg.append(
                f'<rect x="{rx}" y="{ry}" width="{square_in}" height="{square_in}" fill="black"/>'
            )

svg.append('</svg>')

with open("checkerboard_4x6.svg", "w") as f:
    f.write("\n".join(svg))

print("checkerboard_4x6.svg written")
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse("checkerboard_4x6.svg")
root = tree.getroot()

print("SVG size:", root.attrib.get("width"), root.attrib.get("height"))

rects = [e for e in root if e.tag.endswith("rect")]

# first rect is the white background
black = rects[1:]

xs = [float(r.attrib["x"]) for r in black]
ys = [float(r.attrib["y"]) for r in black]
ws = [float(r.attrib["width"]) for r in black]
hs = [float(r.attrib["height"]) for r in black]

print("Square size (in):", ws[0], hs[0])
print("Min x (border left):", min(xs))
print("Min y (border top):", min(ys))

# estimate grid size
unique_x = sorted(set(xs))
unique_y = sorted(set(ys))

print("Squares across:", len(unique_x))
print("Squares down:", len(unique_y))
```
