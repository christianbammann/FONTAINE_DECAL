```python
import cv2
import os

# python script to capture and save images for dataset and calibration

save_dir = "captures"
os.makedirs(save_dir, exist_ok=True)

cap = cv2.VideoCapture(0)  # See3CAM

count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    cv2.imshow("Camera", frame) # show live feed

    key = cv2.waitKey(1) & 0xFF

    if key == ord('s'):  # press 's' to save frame
        filename = os.path.join(save_dir, f"img_{count:03d}.jpg") # saves as img_xxx.jpg
        cv2.imwrite(filename, frame)
        print("Saved:", filename)
        count += 1

    if key == ord('q'):  # press 'q' to quit
        break

cap.release()
cv2.destroyAllWindows()
```
