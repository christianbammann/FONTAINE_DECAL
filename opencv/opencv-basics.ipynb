```python
import cv2
```

```python
img_path = r"C:\Users\cbamm\Downloads\frame_0000.png"
img = cv2.imread(img_path)
scale = 0.33  # adjust 0.3–0.5 as needed
h, w = img.shape[:2]
img = cv2.resize(img, (int(w*scale), int(h*scale)))
```

```python
import numpy as np
```

```python
img_draw = img.copy()
cv2.circle(img_draw, (200, 300), 5, (0, 0, 255), -1)
cv2.line(img_draw, (100, 100), (400, 400), (255, 0, 0), 2)
pts = np.array([[150, 200], [500, 200], [520, 400], [130, 400]], np.int32)
cv2.polylines(img_draw, [pts], isClosed=True, color=(0, 255, 0), thickness=2)
```

```python
cv2.imshow("Drawing Example", img_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert BGR → Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale", gray)
cv2.waitKey(0)

# Convert BGR → HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV", hsv) 
cv2.waitKey(0) 

cv2.destroyAllWindows() # closes all OpenCV windows after you’re done viewing
```

```python
resized = cv2.resize(img, (500, 500))
cv2.imshow("Resized Example", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(gray, (5, 5), 0)
edges = cv2.Canny(blur, 30, 100)

# connect faint lines
edges = cv2.dilate(edges, np.ones((3,3), np.uint8), iterations=1)

# Run Canny edge detection
edges = cv2.Canny(gray, 100, 200)

cv2.imshow("Edges", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Find contours from edge map
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw all contours on a copy of the original image
img_contours = img.copy()
cv2.drawContours(img_contours, contours, -1, (0, 255, 0), 2)

print("Contours found:", len(contours))

cv2.imshow("Contours", img_contours)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# --- Edge Detection ---
blur = cv2.GaussianBlur(gray, (5, 5), 0)          # smooth noise
edges = cv2.Canny(blur, 30, 100)                  # detect edges
edges = cv2.dilate(edges, np.ones((3,3), np.uint8), iterations=1)  # connect broken edges

# --- Contour Detection ---
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# --- Contour Filtering (your code) ---
img_filtered = img.copy()
for c in contours:
    area = cv2.contourArea(c)
    if area > 1000:
        cv2.drawContours(img_filtered, [c], -1, (0, 255, 0), 2)

cv2.imshow("Filtered Contours", img_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
